```python
# vaccine script for BBMP Karnataka : by subhankar chowdhury
import requests
import pandas as pd
import json
import datetime, time
import threading
def slot():
    threading.Timer(3.0, slot).start()
    date = datetime.date.today()
    today = time.strftime("%d-%m-%Y")
    url="https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/findByDistrict?district_id=294&date="+today
    data=requests.get(url).json()
    df=pd.DataFrame.from_dict(data['sessions'])
    columns = ['lat', 'long', 'session_id']
    df.drop(columns, inplace=True, axis=1)
    df1 = df[(df['min_age_limit'] < 45)&(df['available_capacity']>=1)]
    print(df1)
slot()
```
